```python
from pathlib import Path
import shutil
import yaml

# Cấu hình
base_dir = Path("/kaggle/input/bean-dataset-for-yolo-v12")
subdirs = [
    "Frijoles en Buen Etsado.v1i.yolov12",
    "Red anterior.v1i.yolov12",
    "Ya perdi la cuenta.v4i.yolov12",
    "beans.v1i.yolov12",
    "mal estado.v1i.yolov12"
]
splits = ["train", "valid", "test"]

# Thư mục đầu ra
merged_dir = Path("/kaggle/working/merged_yolo_dataset")
for split in splits:
    (merged_dir / "images" / split).mkdir(parents=True, exist_ok=True)
    (merged_dir / "labels" / split).mkdir(parents=True, exist_ok=True)

# Gộp dữ liệu
for sub in subdirs:
    for split in splits:
        img_dir = base_dir / sub / split / "images"
        label_dir = base_dir / sub / split / "labels"
        if not img_dir.exists():
            continue
        for img_file in img_dir.glob("*.*"):
            new_name = f"{sub.replace(' ', '_')}_{img_file.name}"
            shutil.copy(img_file, merged_dir / "images" / split / new_name)
        for label_file in label_dir.glob("*.txt"):
            new_name = f"{sub.replace(' ', '_')}_{label_file.name}"
            shutil.copy(label_file, merged_dir / "labels" / split / new_name)

# Gộp file data.yaml
yaml_template = base_dir / subdirs[0] / "data.yaml"
with open(yaml_template, "r") as f:
    data_yaml = yaml.safe_load(f)

data_yaml["train"] = str((merged_dir / "images/train").as_posix())
data_yaml["val"] = str((merged_dir / "images/valid").as_posix())
data_yaml["test"] = str((merged_dir / "images/test").as_posix())

with open(merged_dir / "data.yaml", "w") as f:
    yaml.dump(data_yaml, f)

print("✅ Đã gộp xong dataset tại:", merged_dir)
```
